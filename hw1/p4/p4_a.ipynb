# Problem 4: Optimal estimation

We have two measurements
$$
Y_i = h_i X + \epsilon_i,\quad i=1,2,
$$
with \(h_1=1,\; h_2=2\). The noise satisfies
$$
\mathbb E[\epsilon_i]=0,\qquad \mathbb E[\epsilon_i^2]=\sigma_i^2,
$$
and \(\epsilon_1,\epsilon_2\) are uncorrelated. Also \( \mathbb E[X\epsilon_i]=0\).

We consider an estimator of the form
$$
\hat X = a_1 Y_1 + a_2 Y_2,
$$
where \(a_1,a_2\) are constants independent of \(X\).

---

## (a)(i) Unbiasedness constraint

First compute the expectation:
$$
\mathbb E[Y_1]=\mathbb E[X+\epsilon_1]=X,\qquad
\mathbb E[Y_2]=\mathbb E[2X+\epsilon_2]=2X.
$$

Thus
$$
\mathbb E[\hat X]=a_1 \mathbb E[Y_1] + a_2 \mathbb E[Y_2]
= a_1 X + a_2 (2X) = (a_1+2a_2)X.
$$

For \(\hat X\) to be unbiased, we require \(\mathbb E[\hat X]=X\) for all \(X\), hence
$$
\boxed{a_1 + 2a_2 = 1.}
$$

---

## (a)(ii) Minimize the MSE \(\mathbb E[(X-\hat X)^2]\)

Write the estimation error:
$$
\begin{aligned}
X-\hat X
&= X - a_1(X+\epsilon_1) - a_2(2X+\epsilon_2) \\
&= \bigl(1-a_1-2a_2\bigr)X - a_1\epsilon_1 - a_2\epsilon_2.
\end{aligned}
$$

Under the unbiasedness constraint \(a_1+2a_2=1\), the \(X\)-term vanishes:
$$
X-\hat X = -a_1\epsilon_1 - a_2\epsilon_2.
$$

Therefore the mean-square error becomes
$$
\begin{aligned}
\mathbb E[(X-\hat X)^2]
&= \mathbb E\bigl[(a_1\epsilon_1 + a_2\epsilon_2)^2\bigr] \\
&= a_1^2\mathbb E[\epsilon_1^2] + a_2^2\mathbb E[\epsilon_2^2]
    + 2a_1a_2\mathbb E[\epsilon_1\epsilon_2].
\end{aligned}
$$

Since \(\epsilon_1,\epsilon_2\) are uncorrelated, \(\mathbb E[\epsilon_1\epsilon_2]=0\). Hence
$$
\boxed{\text{MSE}(a_1,a_2)=a_1^2\sigma_1^2 + a_2^2\sigma_2^2}
\quad \text{subject to } a_1+2a_2=1.
$$

### Lagrange multiplier
Minimize
$$
J(a_1,a_2,\lambda)=a_1^2\sigma_1^2+a_2^2\sigma_2^2+\lambda(a_1+2a_2-1).
$$

Set partial derivatives to zero:
$$
\frac{\partial J}{\partial a_1}=2a_1\sigma_1^2+\lambda=0
\;\Rightarrow\;
a_1=-\frac{\lambda}{2\sigma_1^2},
$$
$$
\frac{\partial J}{\partial a_2}=2a_2\sigma_2^2+2\lambda=0
\;\Rightarrow\;
a_2=-\frac{\lambda}{\sigma_2^2}.
$$

Use the constraint \(a_1+2a_2=1\):
$$
-\frac{\lambda}{2\sigma_1^2} + 2\Bigl(-\frac{\lambda}{\sigma_2^2}\Bigr)=1
\;\Rightarrow\;
-\lambda\Bigl(\frac{1}{2\sigma_1^2}+\frac{2}{\sigma_2^2}\Bigr)=1.
$$

Thus
$$
\lambda=-\frac{1}{\frac{1}{2\sigma_1^2}+\frac{2}{\sigma_2^2}}.
$$

Plugging back gives
$$
a_1=\frac{\frac{1}{2\sigma_1^2}}{\frac{1}{2\sigma_1^2}+\frac{2}{\sigma_2^2}}
=\boxed{\frac{\sigma_2^2}{\sigma_2^2+4\sigma_1^2}},
$$
$$
a_2=\frac{\frac{1}{\sigma_2^2}}{\frac{1}{2\sigma_1^2}+\frac{2}{\sigma_2^2}}
=\boxed{\frac{2\sigma_1^2}{\sigma_2^2+4\sigma_1^2}}.
$$

### Minimum MSE value
$$
\begin{aligned}
\text{MSE}_{\min}
&= a_1^2\sigma_1^2 + a_2^2\sigma_2^2 \\
&= \frac{\sigma_2^4}{(\sigma_2^2+4\sigma_1^2)^2}\sigma_1^2
 + \frac{4\sigma_1^4}{(\sigma_2^2+4\sigma_1^2)^2}\sigma_2^2 \\
&= \frac{\sigma_1^2\sigma_2^2(\sigma_2^2+4\sigma_1^2)}{(\sigma_2^2+4\sigma_1^2)^2}
= \boxed{\frac{\sigma_1^2\sigma_2^2}{\sigma_2^2+4\sigma_1^2}}.
\end{aligned}
$$
