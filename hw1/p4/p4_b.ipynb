## (b) Interpretation in different noise regimes

Recall
$$
a_1=\frac{\sigma_2^2}{\sigma_2^2+4\sigma_1^2},\qquad
a_2=\frac{2\sigma_1^2}{\sigma_2^2+4\sigma_1^2},
\qquad
\hat X=a_1Y_1+a_2Y_2.
$$

### (i) \(\sigma_2^2 \gg \sigma_1^2\)
Then \(\sigma_2^2+4\sigma_1^2 \approx \sigma_2^2\), so
$$
a_1\approx 1,\quad a_2\approx 0.
$$
So \(\hat X \approx Y_1\): we trust the less noisy measurement.

### (ii) \(\sigma_2^2 = \sigma_1^2\)
Then
$$
a_1=\frac{1}{1+4}=\frac{1}{5}=0.2,\qquad
a_2=\frac{2}{1+4}=\frac{2}{5}=0.4.
$$
So \(\hat X=0.2Y_1+0.4Y_2\). Even with equal noise variances, \(Y_2\) has gain \(h_2=2\),
so it effectively provides a strong constraint on \(X\), and thus receives larger weight.

### (iii) \(\sigma_2^2 \ll \sigma_1^2\)
Then \(\sigma_2^2+4\sigma_1^2 \approx 4\sigma_1^2\), so
$$
a_1\approx 0,\quad a_2\approx \frac{1}{2}.
$$
So \(\hat X \approx \frac{1}{2}Y_2\), i.e., we essentially use the high-quality second sensor,
scaled by \(1/h_2\) to estimate \(X\).

These behaviors match intuition: the estimator down-weights the noisier measurement and
relies more on the more reliable one (accounting for the measurement gains \(h_1,h_2\)).
